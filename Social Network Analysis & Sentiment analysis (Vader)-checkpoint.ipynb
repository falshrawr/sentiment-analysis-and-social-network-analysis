```python
#data cleaning and processing 

#ensuring its a string
df['text_ed'] = df['text'].astype(str)

df['text_ed'][3]
```

```python
#remove Urls 
import re

df['text_ed'] = df['text_ed'].apply(lambda x: re.split('https:\/\/.*', str(x))[0])


print(df['text'][26246])
print(df['text_ed'][26246])
```

```python
#removes @usernames from tweets text

def username(tweet):
    Tweet = re.sub('@[^\s]+','',tweet)
    return Tweet

df['text_ed'] = df['text_ed'].apply(username)

print(df['text'][26246])
print(df['text_ed'][26246])
```

```python
#lowercase
df['text_ed'] = df['text_ed'].apply(lambda x: " ".join(x.lower() for x in x.split()))
df['text_ed'][3]
```

```python
punc = string.punctuation 
print(punc)

df['text_ed'] = df['text_ed'].apply(lambda x: " ".join(x for x in x.split() if x not in punc))

df['text_ed'][3]
df
```

```python
#remove stopwords such as he/she/it
stop = stopwords.words('english')

print(stop)

df['text_ed'] = df['text_ed'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))

df['text'][26246]
df['text_ed'][26246]
```

```python
#sentiment analysis on text_ed (cleaned)

analyser = SentimentIntensityAnalyzer()

def sentiment(x):
    score = analyser.polarity_scores(x)
    return score['compound']

df['polarity'] = df['text_ed'].apply(sentiment)
df
```

```python
#add sentiment labels column for polarity rows. 
def label(row):
    if row["polarity"] > 0:
        return ("positive")
    elif row["polarity"] < 0:
        return ("negative")
    else:
        return ("neutral")

df['sentiment'] = df.apply(lambda row: label(row), axis=1)
```

```python
   #CountVectorizer will convert text to a matrix of word token counts for each individual word, 
    #which can be used to calculate the frequencies for each word. 
    #the graph function will then form a graph of the most frequently used words within each tweet subset.

def vectorise(dataframe):
    vector = CountVectorizer(stop_words='english')
    frequency_matrix = vector.fit_transform(dataframe.text_ed)
    sum_frequencies = np.sum(frequency_matrix, axis=0)
    #squeeze removes single-dimensional entries from the shape of an array from applying np.asarray to the sum of frequencies.
    frequency = np.squeeze(np.asarray(sum_frequencies))
    # transform the frequencies for each word into a new dataframe
    frequency_df = pd.DataFrame([frequency], columns=vector.get_feature_names()).transpose()
    return frequency_df

def graph(word_frequency, sent):
    plt.figure(figsize=(10,5))
    plt.bar(np.arange(50), word_frequency[0][1:51], width = 0.8, color = sns.color_palette("muted"), alpha=0.5, 
            edgecolor = "black", capsize=8, linewidth=1);
    labels = word_frequency[0][1:51].index
    plt.xticks(np.arange(50), labels, rotation=90, size=14);
    plt.xlabel("50 more frequent words", size=14);
    plt.ylabel("Frequency", size=14);
    title = "Word Frequency for %s" %sent
    plt.title(title, size=18)
    plt.grid(False);
    plt.gca().spines["top"].set_visible(False);
    plt.gca().spines["right"].set_visible(False);
    plt.show()
    
```

```python
#generate wordcloud visualisation for tweets showing the 100 most common groups of words/phrases in the text

def wordcloud(series, title): 
    wordcloud = WordCloud(width=1600, height=800,max_font_size=100, max_words=100, background_color='white').generate(series)
    plt.figure(figsize=(12,10))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(title, size=12)
    plt.show()

```

```python
#Network Analysis combined with sentiment polarity scores

#first create an edgelist format for dataframe for networkx.
edgelist = pd.DataFrame(df[['screen_name', 'mentions_screen_name','polarity']])
edgelist
```

```python
#first drawing the network graph of whole dataframe for the CSA network

#using a multi directed graph formation
G = nx.MultiDiGraph()
G = nx.from_pandas_edgelist(edgelist, source='screen_name', target='mentions_screen_name', edge_attr='polarity')

#Colour edges by sentiment polarity attribute 
colors = nx.get_edge_attributes(G,'polarity').values()
# red-- yellow -- green colourmap
cmap = plt.cm.RdYlGn

#draws graph including colored edges by polarity
nx.draw(G, node_color='Blue', edge_color=colors, node_size=1, width=2, edge_cmap=cmap, with_labels=False)

# Add colorbar legend and labels
sm = plt.cm.ScalarMappable(cmap=cmap)
cbar = plt.colorbar(sm, ticks=[0,0.5,1])
cbar.ax.set_yticklabels(['-1  Negative','0  Neutral', '1  Positive'])
cbar.ax.set_ylabel('Polarity')

#add title and show graph
plt.title('Social Network graph of CSA network by sentiment polarity score')
plt.figure(figsize=[10,10])
plt.show()
```

```python

#initial measures of graph connectivity
#ensures the network is connected for further analysis and counts how many components form the network 
def component_connectivity(G):
    x = nx.is_connected(G)
    xn = nx.number_connected_components(G)
    return print("the network is connected: {} ,".format(x), '{} components are connected,'.format(xn))
component_connectivity(G)
```

```python
#runs connectivity statistical tests on network graph
def connectivity(G):
    x = nx.is_connected(G)
    xn = nx.number_connected_components(G)
    av = nx.average_clustering(G)
    short = nx.average_shortest_path_length(G)
    dia = nx.diameter(G)
    rad = nx.radius(G)
    tran = nx.transitivity(G)
    n = nx.node_connectivity(G)
    e = nx.edge_connectivity(G)
    return print("the network is connected: {} ,".format(x), '{} components are connected,'.format(xn), 
                 'global clustering coef: {}'.format(av), 'average shortest path length: {}'.format(short),
                 'diameter: {}'.format(dia), 'radius: {}'.format(rad),'transitivity: {} % of triads,'.format(tran),
                 '{} nodes removed to disconnect the network.'.format(n), 
                 '{} edges removed to disconnect the network.'.format(e))
```

```python
#as connectivity measures will not run on an unconnected network, 
#must select the largest component of the network, removing outliers and components of 2 or 3 nodes. 
largest = max(nx.connected_components(G), key=len)
G5 = G.subgraph(largest)
 
connectivity(G5)
```

```python
colors = nx.get_edge_attributes(G5,'polarity').values()
cmap = plt.cm.RdYlGn
nx.draw(G5   , node_color='Blue', edge_color=colors, node_size=1, width=2, edge_cmap=cmap, with_labels=False)

# Add colorbar legend
sm = plt.cm.ScalarMappable(cmap=cmap)
cbar = plt.colorbar(sm, ticks=[0,0.5,1])
cbar.ax.set_yticklabels(['-1 Negative','0 Neutral', '1 Positive'])
cbar.ax.set_ylabel('Polarity')

plt.title('Most Significant Component of the CSA Network Graph by Sentiment Polarity')
plt.figure(figsize=[15,15])
plt.show()
```
